```python
#jhon sebastian hernandez ferreira

from abc import ABC, abstractmethod

#clase base
class Computador:
    def __init__(self, modelo: str, marca: str, procesador: str, memoria_ram_gb: int, almacenamiento_gb: int,) -> None:
        self.modelo = modelo
        self.marca = marca
        self.procesador = procesador
        self.memoria_ram_gb = memoria_ram_gb
        self.almacenamiento_gb = almacenamiento_gb

    @abstractmethod
    def encender(self) -> None:
        pass

## abstraccion inicio + interfaz ##
#clase1
class Computador_gaming(Computador):
    def __init__(self, modelo: str, marca: str, procesador: str, memoria_ram_gb: int, almacenamiento_gb: int, grafica: str,) -> None:
        super().__init__(modelo=modelo, marca=marca, procesador=procesador, memoria_ram_gb=memoria_ram_gb, almacenamiento_gb=almacenamiento_gb)
        self.grafica = grafica

    def encender(self) -> str:
        print("funcionando pero con luces")

#clase2
class computador_uso_cientifico(Computador):

    def __init__(self, modelo: str, marca: str, procesador: str, memoria_ram_gb: int, almacenamiento_gb: int, num_hilos: int, num_procesos: int,) -> None:
        super().__init__(modelo=modelo, marca=marca, procesador=procesador, memoria_ram_gb=memoria_ram_gb, almacenamiento_gb=almacenamiento_gb)
        self.num_hilos = num_hilos
        self.num_procesos = num_procesos

    def encender(self) -> str:
        print("funcionando")

## abstraccion + interfaz final ##
```

```python
## ejemplo de herencia ##
class vehiculo:
    def __init__(self, velocidad_max: int, numero_ruedas: int, marca: str, modelo: str) -> None:
        self.velocidad_max = velocidad_max
        self.numero_ruedas = numero_ruedas
        self.marca = marca
        self.modelo = modelo

    @abstractmethod
    def encender(self) -> None:
        pass

    @abstractmethod
    def andar(self) -> None:
        pass

class moto(vehiculo):
    def __init__(self,velocidad_max: int, numero_ruedas: int, marca: str, modelo: str, cilindrada:int) -> None:
        super().__init__(velocidad_max=velocidad_max, numero_ruedas=numero_ruedas, marca=marca, modelo=modelo)
        self.cilindrada=cilindrada

    def encender(self)->str:
        print("espere 8 segundos")

    def andar(self)->str:
        print(f" la velocidad a la que puede andar la moto es {self.cilindrada-10} km/h")


class carro(vehiculo):
    def __init__(self,velocidad_max: int, numero_ruedas: int, marca: str, modelo: str, traccion:int) -> None:
        super().__init__(velocidad_max=velocidad_max, numero_ruedas=numero_ruedas, marca=marca, modelo=modelo)
        self.traccion=traccion

    def encender(self)->str:
        print("espere 8 segundos")

    def andar(self)->str:
        print(f"puede andar con la fuerda de {self.traccion}")

## fin deejemplo de herencia ##

moto1=moto(200,2,"suzuki","ajs",200)
moto1.andar()



#carro1=
```
